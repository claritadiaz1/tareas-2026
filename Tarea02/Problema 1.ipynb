```python
# Problema 1
# Clasificador de temperatura
# lista desde -30 al 50
lista_numeros = range(-30, 51)
print(lista_numeros)

for x in range(-30, 51):
    if(x < 0):
        print(x, "Muy frío")
    elif(x >= 0 and x <= 24):
        print(x, "Templado")
    else:
        print(x, "Caluroso")
```
